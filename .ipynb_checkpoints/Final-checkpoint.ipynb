```python
#1 WORDS

import nltk
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
from scipy import stats
from nltk.corpus import brown

# Setup Brown Corpus
nltk.download('brown')

# Tokenize the corpus
words = brown.words()

# Calculate word frequencies
word_freq = Counter(words)
sorted_word_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

# Generate an array for ranks
ranks = np.arange(1, len(sorted_word_freq) + 1)

ranks = np.array(ranks)
word_freq = np.array([])

for i in sorted_word_freq:
    word_freq = np.append(word_freq, i[1])

# Create a list of ranks
ranks = np.arange(1, len(word_freq) + 1)

# Linear regression
a, y, r_value, p_value, std_err = stats.linregress(np.log(ranks), np.log(word_freq))
c = np.exp(y)

# Function to calculate the prediction interval
def get_prediction_interval(prediction, y_test, test_predictions, pi=.90):
    sum_errs = np.sum((y_test - test_predictions) ** 2)
    stdev = np.sqrt(1 / (len(y_test) - 2) * sum_errs)
    one_minus_pi = 1 - pi
    ppf_lookup = 1 - (one_minus_pi / 2)
    z_score = stats.norm.ppf(ppf_lookup)
    interval = z_score * stdev
    lower, upper = prediction - interval, prediction + interval
    return lower, prediction, upper

# Calculate the 90% confidence interval for the linear regression
x = np.log(ranks)
y_pred = a * x + y
y_test = np.log(word_freq)
test_predictions = y_pred
pi = 0.90
lower_vet = []
upper_vet = []
for i in range(len(y_test)):
    prediction = y_pred[i]
    lower, prediction, upper = get_prediction_interval(prediction, y_test, test_predictions, pi)
    lower_vet.append(np.exp(lower))
    upper_vet.append(np.exp(upper))
    
# Plot
plt.figure(figsize=(8, 8))
plt.loglog(ranks, word_freq, label='Word Frequencies', marker='o', linestyle='None', markersize=2)
plt.loglog(ranks, np.exp(a * x + y), label='Linear Regression', color='red', linestyle='--')
plt.fill_between(ranks, upper_vet, lower_vet, color='gray', alpha=0.2, label='90% Confidence Interval')
#plt.axvline(x=500, color='green', linestyle='--')
plt.xlabel('log(Rank)')
plt.ylabel('log(Frequency)')
plt.title('Word Frequency Distribution and Linear Fit')
plt.legend()
plt.grid(True)
plt.show()

outside = 0
for i, freq in enumerate(word_freq):
    if freq > upper_vet[i] or freq < lower_vet[i]:
        outside += 1

print("Estimations: c =", c, "a =", a)   
print(f"{outside} / {len(word_freq)} = {outside/len(word_freq)*100}% fall outside confidence bounds")
```

```python
# 2 TRI-GRAM

import nltk
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
from scipy import stats

# Setup Brown Corpus
nltk.download('brown')

# Tokenize the corpus into character-level trigrams
text = "".join(brown.words())  # Combine all words into a single string
tri_grams = [text[i:i+3] for i in range(len(text) - 2)]

# Calculate tri-gram frequencies
tri_gram_freq = Counter(tri_grams)
sorted_tri_gram_freq = sorted(tri_gram_freq.items(), key=lambda x: x[1], reverse=True)

# Generate an array for ranks
ranks = np.arange(1, len(sorted_tri_gram_freq) + 1)

# Create a list of tri-gram frequencies
tri_gram_freq = np.array([i[1] for i in sorted_tri_gram_freq])

# Linear regression
a, y, r_value, p_value, std_err = stats.linregress(np.log(ranks), np.log(tri_gram_freq))
c = np.exp(y)

# Function to calculate the prediction interval
def get_prediction_interval(prediction, y_test, test_predictions, pi=.90):
    sum_errs = np.sum((y_test - test_predictions) ** 2)
    stdev = np.sqrt(1 / (len(y_test) - 2) * sum_errs)
    one_minus_pi = 1 - pi
    ppf_lookup = 1 - (one_minus_pi / 2)
    z_score = stats.norm.ppf(ppf_lookup)
    interval = z_score * stdev
    lower, upper = prediction - interval, prediction + interval
    return lower, prediction, upper

# Calculate the 90% confidence interval for the linear regression
y_test = np.log(tri_gram_freq)
test_predictions = a * np.log(ranks) + y
lower_vet = []
upper_vet = []

for i in range(len(y_test)):
    prediction = test_predictions[i]
    lower, prediction, upper = get_prediction_interval(prediction, y_test, test_predictions)
    lower_vet.append(np.exp(lower))
    upper_vet.append(np.exp(upper))

# Plot
plt.figure(figsize=(8, 8))
plt.loglog(ranks, tri_gram_freq, label='Tri-gram Frequencies', marker='o', linestyle='None', markersize=2)
plt.loglog(ranks, np.exp(a * np.log(ranks) + y), label='Linear Regression', color='red', linestyle='--')
plt.fill_between(ranks, upper_vet, lower_vet, color='gray', alpha=0.2, label='90% Confidence Interval')
plt.xlabel('log(Rank)')
plt.ylabel('log(Frequency)')
plt.title('Tri-gram Frequency Distribution and Linear Fit')
plt.legend()
plt.grid(True)
plt.show()

outside = 0
for i, freq in enumerate(tri_gram_freq):
    if freq > upper_vet[i] or freq < lower_vet[i]:
        outside += 1

print("Estimations: c =", c, "a =", a)   
print(f"{outside} / {len(tri_gram_freq)} = {outside/len(tri_gram_freq)*100}% fall outside confidence bounds")
```

```python
#6

import nltk
import matplotlib.pyplot as plt
from nltk.corpus import brown
from nltk import FreqDist

nltk.download('brown')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# Define the modal verbs
modal_verbs = ['have', 'should', 'shall', 'must', 'will', 'can', 'might']

# Initialize a frequency distribution for part-of-speech tags
pos_freq_dist = FreqDist()

# Iterate through sentences in the Brown Corpus
for sentence in brown.sents():
    # Tokenize the sentence
    tokens = nltk.word_tokenize(' '.join(sentence))
    
    # Find modal verbs in the sentence
    modal_verb_indices = [i for i, token in enumerate(tokens) if token.lower() in modal_verbs]
    
    for index in modal_verb_indices:
        # Extract a window of 2 tokens around the modal verb
        window_start = max(0, index - 2)
        window_end = min(len(tokens), index + 3)  # +3 to include the verb itself

        # Get the words in the window
        window_words = tokens[window_start:window_end]

        # Tag the words in the window with part-of-speech tags
        window_tags = nltk.pos_tag(window_words)

        # Update the frequency distribution
        for word, pos_tag in window_tags:
            pos_freq_dist[pos_tag] += 1

# Extract keys and values from the dictionary
categories = list(pos_freq_dist.keys())
values = list(pos_freq_dist.values())

# Create a bar chart
plt.bar(categories, values)

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart from Dictionary')
plt.xticks(rotation=90)

# Display the chart
plt.show()
```

```python
#8 POLYSEMY

import nltk
from nltk.corpus import brown
from nltk.corpus import wordnet as wn
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
from collections import Counter

# Download the stopwords if you haven't already
nltk.download("stopwords")

short_corpus = [
    word
    for sent in brown.sents()
    if len(sent) <= 20.25
    for word in sent
    if word.lower() not in set(stopwords.words("english"))
]
long_corpus = [
    word
    for sent in brown.sents()
    if len(sent) > 20.25
    for word in sent
    if word.lower() not in set(stopwords.words("english"))
]

def is_homonym(word):
    synsets = wn.synsets(word)
    return len(synsets) > 1

c_short = Counter()
c_long = Counter()

for idx, word in enumerate(short_corpus):
    if is_homonym(word):
        count_around = 0
        for j, w in enumerate(short_corpus[idx - 2 : idx + 3]):
            if j != 2:
                if is_homonym(w):
                    count_around += 1
        c_short[count_around] += 1

for idx, word in enumerate(long_corpus):
    if is_homonym(word):
        count_around = 0
        for j, w in enumerate(long_corpus[idx - 2 : idx + 3]):
            if j != 2:
                if is_homonym(w):
                    count_around += 1
        c_long[count_around] += 1

# Plotting histograms
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(c_short.keys(), c_short.values())
plt.title("Short Sentences")
plt.xlabel("Number of Homonyms Around a Homonym")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
plt.bar(c_long.keys(), c_long.values())
plt.title("Long Sentences")
plt.xlabel("Number of Homonyms Around a Homonym")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
#9 LEXICON

import nltk
from nltk.corpus import brown
from nltk.corpus import wordnet
from collections import Counter, defaultdict
import matplotlib.pyplot as plt

# Step 1: Retrieve the Brown corpus.
nltk.download('brown')
nltk.download('wordnet')

print(len(brown.sents()))  # Print the number of sentences in the corpus

# Step 3: Create short and long sentence corpora.
short_sentences = [s for s in brown.sents() if len(s) <= 20]
long_sentences = [s for s in brown.sents() if len(s) > 20]

print(len(short_sentences))
print(len(long_sentences))

with open('lexicon.tff', 'r') as tff_file:
    tff_data = tff_file.read().split('\n')

lexicon = []  # Initialize an empty list to store lexicon data

for line in tff_data:
    # Split the line into key-value pairs
    if line != " ":
        pairs = line.strip().split(" ")

        word_data = {}
        for pair in pairs:
            key, value = pair.split('=')
            word_data[key] = value
            lexicon.append(word_data)

long_tokens = []

for sentence in long_sentences:
    for token in sentence:
        long_tokens.append(token)

short_tokens = []

for sentence in short_sentences:
    for token in sentence:
        short_tokens.append(token)

positive_words = 0
negative_words = 0

for i, token in enumerate(long_tokens):
    print(i, "/", len(long_tokens))
    for word_data in lexicon:
        if word_data["word1"] == token:
            print(token)
            if word_data["priorpolarity"] == "negative":
                negative_words += 1
            else:
                positive_words += 1
                
# Print the positive and negative words
print("Positive Words:", positive_words)
print("Negative Words:", negative_words)
```
