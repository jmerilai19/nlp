```python
#1 Zipf's for words

import nltk
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
from scipy import stats

# Setup Brown Corpus
nltk.download('brown')
from nltk.corpus import brown

# Tokenize the corpus
words = brown.words()

# Calculate word frequencies
word_freq = Counter(words)
sorted_word_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

# Create a list of ranks
ranks = np.arange(1, len(sorted_word_freq) + 1)

# Step 8: Plot the log-scale of frequency versus ranking
plt.figure(figsize=(8, 8))
plt.loglog(ranks, [freq for _, freq in sorted_word_freq], label="Word Frequencies")

# Fit a linear regression line
coefficients = np.polyfit(np.log(ranks), np.log([freq for _, freq in sorted_word_freq]), 1)
linear_regression = np.poly1d(coefficients)

# Calculate the best-fit 'a' parameter
a = coefficients[0]

# Calculate 'C' using Zipf's law
f1 = sorted_word_freq[0][1]  # Frequency of the most frequent word (rank 1)
C = f1 / (1**a)

# Plot the linear regression line
x_range = ranks  # Use ranks to match the size
plt.loglog(x_range, np.exp(linear_regression(np.log(x_range))), 'g:', linestyle='--')

# Calculate the 90% confidence bounds
n = len(ranks)
p = 0.9
t_value = abs(stats.t.ppf((1 - (1 - p) / 2), n - 2))
y_hat = linear_regression(np.log(x_range))
sse = np.sum((np.log([freq for _, freq in sorted_word_freq]) - y_hat) ** 2)
stdev = np.sqrt(1 / (n - 2) * sse)
conf_int = t_value * stdev * np.sqrt(1 + 1/n + (np.log(x_range) - np.mean(np.log(x_range))) ** 2)

# Upper and lower confidence bounds
upper_bound = np.exp(y_hat + conf_int)
lower_bound = np.exp(y_hat - conf_int)

# Calculate the actual word frequencies for each word in the Brown Corpus
actual_word_frequencies = [freq for _, freq in sorted_word_freq]

# Count the number of tokens outside the 90% confidence bounds
outside_bounds = 0
for actual_freq, upper, lower in zip(actual_word_frequencies, upper_bound, lower_bound):
    if actual_freq < lower or actual_freq > upper:
        outside_bounds += 1

# Calculate the total number of tokens
total_tokens = sum(actual_word_frequencies)

# Calculate the percentage of tokens outside the 90% confidence bounds
percentage_outside_bounds = (outside_bounds / total_tokens) * 100
print(f"Percentage of tokens outside the 90% confidence bounds: {percentage_outside_bounds:.2f}%")

# Fill the 90% confidence bounds with a gray color
plt.fill_between(x_range, lower_bound, upper_bound, color='gray', alpha=0.1, label='90% Confidence Bounds')

plt.xlabel('Rank (log-scale)')
plt.ylabel('Frequency (log-scale)')
plt.legend()
plt.title('Word Frequency in the Brown Corpus')
plt.show()

# Output the estimated 'C' and 'a'
print(f"Estimated C: {C:.2f}")
print(f"Estimated a: {a:.2f}")
```

```python
import nltk
from nltk.corpus import brown
from nltk.corpus import stopwords
import numpy as np
from scipy.stats import describe

nltk.download('brown')
nltk.download('stopwords')

# Load the Brown Corpus and stopwords
brown_corpus = brown.sents()
stop_words = set(stopwords.words('english'))

# Function to count stopwords in a sentence
def count_stopwords(sentence):
    return sum(1 for word in sentence if word.lower() in stop_words)

# Calculate the number of stopwords per sentence and store in a list
stopword_counts = [count_stopwords(sentence) for sentence in brown_corpus]

# Calculate mean, standard deviation, and kurtosis
mean_stopwords = np.mean(stopword_counts)
std_dev_stopwords = np.std(stopword_counts)
kurtosis_stopwords = describe(stopword_counts).kurtosis

# Print the results
print(f"Mean number of stopwords per sentence: {mean_stopwords}")
print(f"Standard deviation of stopwords per sentence: {std_dev_stopwords}")
print(f"Kurtosis value of stopwords per sentence: {kurtosis_stopwords}")
```

```python
# Modified Zipf's for words considering stopwords per sentence

import nltk
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
from scipy import stats
from nltk.corpus import brown
from nltk.corpus import stopwords

# Download the Brown Corpus and stopwords
nltk.download('brown')
nltk.download('stopwords')

# Tokenize the Brown Corpus and get stopwords
brown_corpus = brown.sents()
stop_words = set(stopwords.words('english'))

# Calculate word frequencies, considering stopwords per sentence
word_freq = Counter()
sentence_stopword_freq = Counter()

for sentence in brown_corpus:
    sentence_word_freq = Counter(sentence)
    sentence_stopword_freq = Counter([word for word in sentence if word.lower() in stop_words])
    
    for word, freq in sentence_word_freq.items():
        if word not in stop_words:
            word_freq[word] += freq - sentence_stopword_freq[word]
            
# Sort the word frequencies
sorted_word_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

# Create a list of ranks
ranks = np.arange(1, len(sorted_word_freq) + 1)

# Step 8: Plot the log-scale of frequency versus ranking
plt.figure(figsize=(8, 8))
plt.loglog(ranks, [freq for _, freq in sorted_word_freq], label="Word Frequencies")

# Fit a linear regression line
coefficients = np.polyfit(np.log(ranks), np.log([freq for _, freq in sorted_word_freq]), 1)
linear_regression = np.poly1d(coefficients)

# Calculate the best-fit 'a' parameter
a = coefficients[0]

# Calculate 'C' using Zipf's law
f1 = sorted_word_freq[0][1]  # Frequency of the most frequent word (rank 1)
C = f1 / (1**a)

# Plot the linear regression line
x_range = ranks  # Use ranks to match the size
plt.loglog(x_range, np.exp(linear_regression(np.log(x_range))), 'g:', linestyle='--')

# Calculate the 90% confidence bounds
n = len(ranks)
p = 0.9
t_value = abs(stats.t.ppf((1 - (1 - p) / 2), n - 2))
y_hat = linear_regression(np.log(x_range))
sse = np.sum((np.log([freq for _, freq in sorted_word_freq]) - y_hat) ** 2)
stdev = np.sqrt(1 / (n - 2) * sse)
conf_int = t_value * stdev * np.sqrt(1 + 1/n + (np.log(x_range) - np.mean(np.log(x_range))) ** 2)

# Upper and lower confidence bounds
upper_bound = np.exp(y_hat + conf_int)
lower_bound = np.exp(y_hat - conf_int)

# Calculate the actual word frequencies for each word
actual_word_frequencies = [freq for _, freq in sorted_word_freq]

# Count the number of tokens outside the 90% confidence bounds
outside_bounds = 0
for actual_freq, upper, lower in zip(actual_word_frequencies, upper_bound, lower_bound):
    if actual_freq < lower or actual_freq > upper:
        outside_bounds += 1

# Calculate the total number of tokens
total_tokens = sum(actual_word_frequencies)

# Calculate the percentage of tokens outside the 90% confidence bounds
percentage_outside_bounds = (outside_bounds / total_tokens) * 100
print(f"Percentage of tokens outside the 90% confidence bounds: {percentage_outside_bounds:.2f}%")

# Fill the 90% confidence bounds with a gray color
plt.fill_between(x_range, lower_bound, upper_bound, color='gray', alpha=0.1, label='90% Confidence Bounds')

plt.xlabel('Rank (log-scale)')
plt.ylabel('Frequency (log-scale)')
plt.legend()
plt.title('Word Frequency in the Brown Corpus (with Stopwords per Sentence)')
plt.show()

# Output the estimated 'C' and 'a'
print(f"Estimated C: {C:.2f}")
print(f"Estimated a: {a:.2f}")
```

```python
#6

import nltk
import matplotlib.pyplot as plt
from nltk.corpus import brown
from nltk import FreqDist

nltk.download('brown')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# Define the modal verbs
modal_verbs = ['have', 'should', 'shall', 'must', 'will', 'can', 'might']

# Initialize a frequency distribution for part-of-speech tags
pos_freq_dist = FreqDist()

# Iterate through sentences in the Brown Corpus
for sentence in brown.sents():
    # Tokenize the sentence
    tokens = nltk.word_tokenize(' '.join(sentence))
    
    # Find modal verbs in the sentence
    modal_verb_indices = [i for i, token in enumerate(tokens) if token.lower() in modal_verbs]
    
    for index in modal_verb_indices:
        # Extract a window of 2 tokens around the modal verb
        window_start = max(0, index - 2)
        window_end = min(len(tokens), index + 3)  # +3 to include the verb itself

        # Get the words in the window
        window_words = tokens[window_start:window_end]

        # Tag the words in the window with part-of-speech tags
        window_tags = nltk.pos_tag(window_words)

        # Update the frequency distribution
        for word, pos_tag in window_tags:
            pos_freq_dist[pos_tag] += 1

# Plot the distribution of part-of-speech tags
plt.figure(figsize=(12, 6))
pos_freq_dist.plot(30, title='Distribution of Part-of-Speech Tags Around Modal Verbs')
plt.show()
```

```python
#7

import nltk
import matplotlib.pyplot as plt
from collections import Counter

# Download the Brown Corpus if not already available
nltk.download('brown')

# Load the Brown Corpus
from nltk.corpus import brown
brown_corpus = brown.sents()

# Calculate the average sentence length
sentence_lengths = [len(sentence) for sentence in brown_corpus]
average_length = sum(sentence_lengths) / len(sentence_lengths)

# Create subsets of sentences based on length
long_sentences = [sentence for sentence in brown_corpus if len(sentence) > average_length]
short_sentences = [sentence for sentence in brown_corpus if len(sentence) <= average_length]

# Tokenize the sentences
def tokenize_sentences(sentences):
    tokens = [word for sentence in sentences for word in sentence]
    return tokens

long_tokens = tokenize_sentences(long_sentences)
short_tokens = tokenize_sentences(short_sentences)

# Count token frequencies
long_token_counts = Counter(long_tokens)
short_token_counts = Counter(short_tokens)

# Sort token frequencies in descending order
long_token_counts = dict(sorted(long_token_counts.items(), key=lambda item: item[1], reverse=True))
short_token_counts = dict(sorted(short_token_counts.items(), key=lambda item: item[1], reverse=True))

# Plot Zipfian distribution for both subsets
def plot_zipf_distribution(token_counts, title):
    rank = range(1, len(token_counts) + 1)
    counts = list(token_counts.values())
    plt.plot(rank, counts)
    plt.xlabel('Rank')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.xscale('log')
    plt.yscale('log')
    plt.show()

plot_zipf_distribution(long_token_counts, 'Zipfian Distribution (Long Sentences)')
plot_zipf_distribution(short_token_counts, 'Zipfian Distribution (Short Sentences)')
```

```python
import nltk
from nltk.corpus import brown
from nltk.corpus import wordnet
from collections import defaultdict
import matplotlib.pyplot as plt

# Step 1: Retrieve the Brown corpus.
nltk.download('brown')
nltk.download('wordnet')

# Step 2: Create a function to identify tokens with more than one sense.
def get_polysemous_tokens(corpus, lexical_db):
    polysemous_tokens = set()
    for sentence in corpus:
        for word in sentence:
            synsets = lexical_db.synsets(word)
            if len(synsets) > 1:
                polysemous_tokens.add(word)
    return polysemous_tokens

# Step 3: Create short and long sentence corpora.
short_sentences = [s for s in brown.sents() if len(s) <= 10]
long_sentences = [s for s in brown.sents() if len(s) > 10]

# Step 4-5: Implement a script to count tokens with multiple senses in a 5-token interval.
def count_polysemous_in_interval(sentence, lexical_db, interval_size=5):
    count = defaultdict(int)
    for i in range(len(sentence)):
        word = sentence[i]
        if word in polysemous_tokens:
            left_context = sentence[max(i - interval_size, 0):i]
            right_context = sentence[i + 1:i + interval_size + 1]
            context = left_context + right_context
            for context_word in context:
                if lexical_db.synsets(context_word) and len(lexical_db.synsets(context_word)) > 1:
                    count[word] += 1
    return count

# Step 6: Visualize a histogram.
def plot_histogram(data, title, xlabel, ylabel):
    plt.hist(data, bins=range(max(data) + 2), align='left', rwidth=0.8)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

# Step 7: Plot a histogram of the part-of-speech tags of homonyms.
def plot_pos_histogram(polysemous_tokens, lexical_db):
    pos_counts = defaultdict(int)
    for word in polysemous_tokens:
        synsets = lexical_db.synsets(word)
        for synset in synsets:
            pos_counts[synset.pos()] += 1
    plt.bar(pos_counts.keys(), pos_counts.values())
    plt.title("Part-of-Speech Distribution of Homonyms")
    plt.xlabel("Part of Speech")
    plt.ylabel("Frequency")
    plt.show()

# Find polysemous tokens in the Brown corpus.
lexical_db = wordnet
polysemous_tokens = get_polysemous_tokens(brown.words(), lexical_db)

# Process short and long sentence corpora and count the context tokens.
short_sentence_counts = defaultdict(int)
long_sentence_counts = defaultdict(int)

for sentence in short_sentences:
    counts = count_polysemous_in_interval(sentence, lexical_db)
    for word, count in counts.items():
        short_sentence_counts[word] += count

for sentence in long_sentences:
    counts = count_polysemous_in_interval(sentence, lexical_db)
    for word, count in counts.items():
        long_sentence_counts[word] += count

# Now you can plot the histograms as before.
plot_histogram(short_sentence_counts.values(), "Short Sentence Corpus", "Number of Other Homonyms in Interval", "Frequency")
plot_histogram(long_sentence_counts.values(), "Long Sentence Corpus", "Number of Other Homonyms in Interval", "Frequency")

# Plot a histogram of the part-of-speech tags of homonyms.
plot_pos_histogram(polysemous_tokens, lexical_db)
```

```python
import nltk
from nltk.corpus import brown
from nltk.corpus import wordnet
from collections import defaultdict
import matplotlib.pyplot as plt

# Step 1: Retrieve the Brown corpus.
nltk.download('brown')
nltk.download('wordnet')

# Step 3: Create short and long sentence corpora.
short_sentences = [s for s in brown.sents() if len(s) <= 10]
long_sentences = [s for s in brown.sents() if len(s) > 10]

with open('lexicon.tff', 'r') as tff_file:
    tff_data = tff_file.read().split('\n')

lexicon = []  # Initialize an empty list to store lexicon data

for line in tff_data:
    # Split the line into key-value pairs
    if line != " ":
        pairs = line.strip().split(" ")

        word_data = {}
        for pair in pairs:
            key, value = pair.split('=')
            word_data[key] = value
            lexicon.append(word_data)
            
# Tokenize the sentences
def tokenize_sentences(sentences):
    tokens = [word for sentence in sentences for word in sentence]
    return tokens

long_tokens = tokenize_sentences(long_sentences)
short_tokens = tokenize_sentences(short_sentences)

positive_words = 0
negative_words = 0

for i, token in enumerate(long_tokens):
    print(i, "/", len(long_tokens))
    for word_data in lexicon:
        if word_data["word1"] == token:
            print(token)
            if word_data["priorpolarity"] == "negative":
                negative_words += 1
            else:
                positive_words += 1
                
# Print the positive and negative words
print("Positive Words:", positive_words)
print("Negative Words:", negative_words)
```

```python
import nltk
from nltk.corpus import brown
from nltk.corpus import wordnet
from collections import Counter, defaultdict
import matplotlib.pyplot as plt

# Step 1: Retrieve the Brown corpus.
nltk.download('brown')
nltk.download('wordnet')

# Step 3: Create short and long sentence corpora.
short_sentences = [s for s in brown.sents() if len(s) <= 20]
long_sentences = [s for s in brown.sents() if len(s) > 20]

with open('lexicon.tff', 'r') as tff_file:
    tff_data = tff_file.read().split('\n')

lexicon = []  # Initialize an empty list to store lexicon data

for line in tff_data:
    # Split the line into key-value pairs
    if line != " ":
        pairs = line.strip().split(" ")

        word_data = {}
        for pair in pairs:
            key, value = pair.split('=')
            word_data[key] = value
            lexicon.append(word_data)
            
# Tokenize the sentences
def tokenize_sentences(sentences):
    tokens = [word for sentence in sentences for word in sentence]
    return tokens

long_tokens = tokenize_sentences(long_sentences)
short_tokens = tokenize_sentences(short_sentences)

long_tokens = Counter(long_tokens)
short_tokens = Counter(short_tokens)

positive_words = 0
negative_words = 0

for i, token in enumerate(long_tokens):
    print(i, "/", len(long_tokens))
    for word_data in lexicon:
        if word_data["word1"] == token:
            print(token[0])
            if word_data["priorpolarity"] == "negative":
                negative_words += token[1]
            else:
                positive_words += token[1]
                
# Print the positive and negative words
print("Positive Words:", positive_words)
print("Negative Words:", negative_words)
```
